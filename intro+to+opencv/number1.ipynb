```python
# Create a black image
# Create white background
img = np.ones((500, 500, 3), dtype=np.uint8) * 255

# Draw full yellow circle (Pac-Man's body)
cv.circle(img, (250, 250), 200, (255, 255, ), -1)  # Yellow

# Draw triangle to create the mouth (cut out a wedge)
pts = np.array([[250, 250], [450, 150], [450, 350]], np.int32)
pts = pts.reshape((-1, 1, 2))
cv.fillPoly(img, [pts], (255, 255, 255))  # Fill triangle with background color (white)

# Draw eye (small black circle)
cv.circle(img, (250, 150), 20, (0, 0, 0), -1)  # Black eye

# Show the result
imshow("Pac-Man (Primitive Shapes)", img)


```

```python
cv.putText(img, 'Hello, my name is PAC!!!', (160,225), cv.FONT_HERSHEY_TRIPLEX, .8, (0,255,0), 2)
imshow('Text', img)
```
